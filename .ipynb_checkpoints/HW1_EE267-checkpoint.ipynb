```python
#Calls for importing all the libraries used in the file 
from skimage.io import imread, imshow #consists of functions useful to grayscale image on local device
from skimage import data #Imports images from skimage

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
image_color = imread ('OriginalLena.png', as_gray=False) #reads image in local repo through name and is assigned to image_color
                                                         #Has option to apply grayscale, but does not apply in this case
print (image_color.shape) #prints out the pixel length, width, and color channels of the image in terms of comma separated values
                          #4 represent the three primary color intensity (RGB) and the last channel is for transparency
imshow(image_color) #prints out image assinged to image_color
```

# Question 2

```python


# Read the image and convert it to grayscale
image = cv2.imread('OriginalLena.png', cv2.IMREAD_GRAYSCALE)

# Define Prewitt kernels
prewitt_x = np.array([[-1, 0, 1], 
                       [-1, 0, 1], 
                       [-1, 0, 1]], dtype=np.float32)

prewitt_y = np.array([[-1, -1, -1], 
                       [0,  0,  0], 
                       [1,  1,  1]], dtype=np.float32)

# Define Sobel kernels
sobel_x = np.array([[-1, 0, 1], 
                     [-2, 0, 2], 
                     [-1, 0, 1]], dtype=np.float32)

sobel_y = np.array([[-1, -2, -1], 
                     [0,  0,  0], 
                     [1,  2,  1]], dtype=np.float32)

# Apply convolution using OpenCV's filter2D function
Gx_prewitt = cv2.filter2D(image, cv2.CV_64F, prewitt_x)
Gy_prewitt = cv2.filter2D(image, cv2.CV_64F, prewitt_y)
G_prewitt = np.sqrt(Gx_prewitt**2 + Gy_prewitt**2)

Gx_sobel = cv2.filter2D(image, cv2.CV_64F, sobel_x)
Gy_sobel = cv2.filter2D(image, cv2.CV_64F, sobel_y)
G_sobel = np.sqrt(Gx_sobel**2 + Gy_sobel**2)

# Display results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(G_prewitt, cmap='gray')
plt.title('Prewitt Edge Detection')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(G_sobel, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```
